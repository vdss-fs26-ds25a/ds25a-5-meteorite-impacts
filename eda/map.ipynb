```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scattermapbox(
    lat=[33.725],
    lon=[130.75],
    mode="markers",
    marker=dict(size=10
                , color="#eb1717"),
    text=["Nogata"]
))

fig.update_layout(
    width=1100,
    height=800,  # 👈 makes map much bigger for screenshots

    mapbox=dict(
        style="white-bg",
        zoom=6,  # 👈 slightly closer = sharper perceived detail
        center=dict(lat=33.725, lon=130.75),

        layers=[
            # 🛰️ satellite
            dict(
                sourcetype="raster",
                source=[
                    "https://services.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}"
                ],
                below="traces"
            ),
            # 🏷️ labels
            dict(
                sourcetype="raster",
                source=[
                    "https://services.arcgisonline.com/ArcGIS/rest/services/Reference/World_Boundaries_and_Places/MapServer/tile/{z}/{y}/{x}"
                ],
                below="traces"
            )
        ]
    ),

    margin=dict(r=0, t=0, l=0, b=0)
)

fig.show()
```
